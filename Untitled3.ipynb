## Step 1 : Load Data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder, MinMaxScaler
from sklearn.cluster import KMeans
from scipy.stats import skew
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
```

```python
# Get an overview of the dataset
df.info()
```

**Why?**

## Step 3 : Drop Unnecessary Columns 

```python
# Remove columns with more than 70% missing values
df2 = df[[column for column in df if df[column].count() / len(df) >= 0.3]]  # Keep columns with at least 30% non-null values

# Drop the Id column
del df2['Id']

print("Dropped columns:", [c for c in df.columns if c not in df2.columns])
```

```python
# Update df
df = df2
```

**Why?**
- `Id` is just an **identifier** and doesn't help in predicting house prices.
- Columns with **too many missing values** (like `Alley`, `PoolQC`) don't contribute useful insights.

## Step 4 : Visualize Sales Price Distribution 

```python
# Describe SalePrice column
print(df['SalePrice'].describe())
```

```python
# Plot distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['SalePrice'], bins=100, kde=True, color="g")
plt.title("Distribution of Sale Prices")
plt.show()
```

**Why?**
- To check if **SalePrice** follows a normal distribution or if it's skewed.
- Helps in deciding if **log transformation** is needed for better modeling.

With this information we can see that prices are skewed to the right and some extreme values â€‹â€‹are above ~500,000. Finally, we'll want to get rid of these to get the normal distribution of the argument ("SalePrice") 

## Step 5 : Handling Missing 

```python
# Check missing values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print(missing_values.sort_values(ascending=False))
```

```python
# Fill missing numerical values with median for robustness
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing categorical values with the mode
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

# Verify missing values after handling
print("Missing values after handling:", df.isnull().sum().sum())
```

**Why?**
- Identifies which columns have **missing values**.
- Helps decide whether to **drop** or **impute** them.

## Step 7 : Find Correlations

```python
# Compute correlation with SalePrice
df_corr = df_num.corr()['SalePrice'].sort_values(ascending=False)

# Select highly correlated features
golden_features = df_corr[abs(df_corr) > 0.5]
print("Highly correlated features:")
print(golden_features)
```

## Step 9 :  Handle Outliers 

```python
# Remove outliers in GrLivArea
df = df[df['GrLivArea'] < 4000]
```

**Why?**
- Outliers can **skew** correlation results and affect model accuracy.

## Step 10:  Feature Engineering

```python
# Create a new feature indicating presence of a basement
df['HasBasement'] = df['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
```

**Why?**
- Helps capture **binary information** (presence vs. absence of a feature).
- Improves model **interpretability**.

## Step 11 : Encode Categorical Data

```python
# Convert categorical variables into dummy variables
df = pd.get_dummies(df, drop_first=True)
```

**Why?**
- Machine learning models require **numerical inputs**.
- Avoids misinterpretation of **categorical variables**.
